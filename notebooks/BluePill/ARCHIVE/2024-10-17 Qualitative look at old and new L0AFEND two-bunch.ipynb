```python
csrTF = False
tao = initializeTao(
    inputBeamFilePathSuffix = '/beams/nmmToL0AFEND_2bunch_2024-02-16Clean/2024-02-16_2bunch_1e5Downsample_nudgeWeights.h5',
    #inputBeamFilePathSuffix = '/beams/2024-10-14_Impact_TwoBunch/2024-10-14_TwoBunch.h5',
    
    csrTF = csrTF,
    numMacroParticles=1e4,
)

tao.cmd("set ele TCY10490 VOLTAGE = 1.06e6")

trackBeam(tao, trackEnd = "BEGBC11_1", verbose = True)
POld = ParticleGroup(data=tao.bunch_data("L0BFEND"))
display(plotMod(POld, 'delta_t', 'pz', bins=300))
display(plotMod(POld, 'x', 'y', bins=300))

POld = ParticleGroup(data=tao.bunch_data("PR10711"))
display(plotMod(POld, 'x', 'y', bins=300))

plt.clf()
```

```python
# Create a 2D density histogram with 20x20 bins
plt.hist2d(POld.x, POld.y, bins=[100, 100])

# Set aspect ratio to 'auto' to match Mathematica's behavior
plt.gca().set_aspect('equal')

# Display the plot
plt.show()
```
