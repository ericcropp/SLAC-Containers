```python
from UTILITY_quickstart import *

with open('setLattice_defaults.yml', 'r') as file:
    importedDefaultSettings = yaml.safe_load(file)
```

```python
tao = initializeTao(
    inputBeamFilePathSuffix = '/beams/2024-12-11_Impact_OneBunch/2024-12-11_oneBunch.h5',
    csrTF = True,
    #numMacroParticles=1e4,
)

trackEnd = "ENDBC14_2" 

trackBeam(tao, 
          trackEnd = trackEnd,
          centerBC14 = True, 
          #verbose = True
         )
```

## Initial beam

```python
PInitial = getBeamAtElement(tao, "L0AFEND")

display(plotMod(PInitial, 'x', 'xp',  bins=300))
plt.clf()


print(PInitial)
print(f"""Emittance: {emittance(PInitial, plane = "x", fraction = 0.9)}, {emittance(PInitial, plane = "y", fraction = 0.9)} m-rad""")
```

## Into BC14

```python
PInput = getBeamAtElement(tao, "BEGBC14_1")

display(plotMod(PInput, 'x', 'xp',  bins=300))
display(plotMod(PInput, 'x', 'energy',  bins=300))


print(PInput)
print(f"""Emittance: {emittance(PInput, plane = "x", fraction = 0.9)}, {emittance(PInput, plane = "y", fraction = 0.9)} m-rad""")
```

## Out of BC14

```python
POutput = getBeamAtElement(tao, trackEnd)

display(plotMod(POutput, 'x', 'xp',  bins=300))
display(plotMod(POutput, 'x', 'energy',  bins=300))


print(POutput)
print(f"""Emittance: {emittance(POutput, plane = "x", fraction = 0.9)}, {emittance(POutput, plane = "y", fraction = 0.9)} m-rad""")
```
