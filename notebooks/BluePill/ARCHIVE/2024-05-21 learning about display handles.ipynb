```python
import matplotlib.pyplot as plt
from IPython.display import display, clear_output, update_display
import time
from IPython.display import DisplayHandle

class DemoOptimizer:
    def __init__(self):
        self.totalNumEvals = 0
        self.plot_display_handle = None
        self.evals_display_handle = None

    def update_plot(self):
        plt.figure()
        plt.plot([1, 2, 3], [1, 4, 9])
        plt.title(f"Total Num Evals: {self.totalNumEvals}")
        
        if self.plot_display_handle is None:
            self.plot_display_handle = display(plt.gcf(), display_id=True)
        else:
            update_display(plt.gcf(), display_id=self.plot_display_handle.display_id)
        plt.close()

    def display_evals(self):
        if self.evals_display_handle is None:
            self.evals_display_handle = display(f"Total Num Evals: {self.totalNumEvals}", display_id=True)
        else:
            update_display(f"Total Num Evals: {self.totalNumEvals}", display_id=self.evals_display_handle.display_id)

    def specificOptimizer(self):
        self.totalNumEvals += 1
        
        # Update only the specific evals output
        self.display_evals()
        
        # For demonstration, we'll update the plot every 5 evaluations
        if self.totalNumEvals % 5 == 0:
            self.update_plot()

    def run(self):
        for _ in range(20):  # Simulate 20 iterations
            self.specificOptimizer()
            time.sleep(0.5)  # Shorter sleep for quicker demonstration

optimizer = DemoOptimizer()
optimizer.run()
```
