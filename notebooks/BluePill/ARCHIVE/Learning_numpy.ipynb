```python
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20, r=2):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much

    return divtime
```

```python
plt.clf()
plt.imshow(mandelbrot(500, 500))
```

```python
rg = np.random.default_rng(1)
v = rg.normal(0,1, int(1e6))
plt.hist(v, bins = 5000, density = True);

del v
```

```python
v = rg.normal(0,1, int(1e6))
w = rg.normal(0,2, len(v))

plt.figure(figsize=(6,6))
#plt.hist2d(v,w,bins=100,range=[[-5,5],[-5,5]]);
plt.hexbin(
    v, 
    w,
    gridsize = 50,
    extent = (-5, 5, -5, 5)
);
```

```python
# I asked: "I'm still trying to learn how to do things in a matplotlib-like way instead of Mathematica. For example, in Mathematica I'd generate a 2D array of x and y values then map a rotation matrix onto each element to rotate it. In python, it seems the preference is to have separate lists for x and y. Is that true in general? If so, what's the preferred way to perform this rotation?"
# ChatGPT suggests the following approach (combining the two vectors into a 2D array, do math, then split back into vectors) is the most "Pythonic"
# I don't love it, but I suppose I can stomach it

# Angle of rotation in degrees
theta = np.radians(45)  # Convert to radians

# Rotation matrix
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# Apply the rotation to each point
vw = np.vstack((v, w))  # Stack X and Y coordinates
rotated_vw = np.dot(rotation_matrix, vw)  # Matrix multiplication

# Extract rotated coordinates
rotated_v, rotated_w = rotated_vw

plt.figure(figsize=(6,6))
plt.hexbin(
    rotated_v, 
    rotated_w,
    gridsize = 50,
    extent = (-5, 5, -5, 5)
);
```

```python
#With enough arm twisting, ChatGPT admits that having a named list for each dimension could be a bad solution if the dimension is higher.
#Here's an alternative, more Mathematica-like approach to elementwise transforms.

vw = np.column_stack((v,w))
```

```python
vw
```

```python
transformed_vw = np.zeros_like(vw)
#transformed_vw[:,0] = vw[:,0]**2
#transformed_vw[:,1] = 2*vw[:,1] + vw[:,0]

transformed_vw[:,0] = vw[:,0] - 0.25*vw[:,1]**3 
transformed_vw[:,1] = vw[:,1]
```

```python
plt.figure(figsize=(6,6))
plt.hexbin(
    transformed_vw[:, 0], 
    transformed_vw[:, 1],
    gridsize = 100,
    extent = (-5, 5, -5, 5)
);
```

```python
#Here's another method that uses an external function
def foo(point):
    x, y = point  # Unpack the point into x and y
    return [x + 0.25 * y**3, y]


transformed_vw = np.apply_along_axis(foo, 1, vw)

plt.figure(figsize=(6,6))
plt.hexbin(
    transformed_vw[:, 0], 
    transformed_vw[:, 1],
    gridsize = 100,
    extent = (-5, 5, -5, 5)
);
```

```python
#And this is the closest yet to the way I'd do it in Mathematica; using lambda functions
transformed_vw = np.apply_along_axis(
    lambda point: [point[0] - 0.25 * point[1]**3, point[1]], 
    1, 
    vw)

plt.figure(figsize=(6,6))
plt.hexbin(
    transformed_vw[:, 0], 
    transformed_vw[:, 1],
    gridsize = 100,
    extent = (-5, 5, -5, 5)
);
```

```python
#Just for fun, a more complex transform using some more numpy functions
def foo(point):
    x, y = point  # Unpack the point into x and y
    #theta = 3 / np.linalg.norm(point)
    theta = 1.2 * np.linalg.norm(point)

    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])
    
    return np.dot(rotation_matrix, point)


transformed_vw = np.apply_along_axis(foo, 1, vw)

plt.figure(figsize=(6,6))
plt.hexbin(
    transformed_vw[:, 0], 
    transformed_vw[:, 1],
    gridsize = 100,
    extent = (-5, 5, -5, 5)
);
```
