```python
def makeDistgenFile(
    MTESet = 400,
    sigmaSet = 1.7,
    cutSet = 2.7,
    VCCFileNameSet = "VCC"
):
    # Define the input and output filenames
    input_filename = "./impact/2024-10-07_distgen_template.yaml"
    output_filename = "./impact/2024-10-07_distgen.yaml"
    
    # Define the new value to replace "MTESet" with
    new_value = "YOUR_NEW_VALUE"  # Replace this with the actual value you want to use
    
    # Read the template file
    with open(input_filename, 'r') as file:
        content = file.read()
    

    content = content.replace("MTESet", str(MTESet))
    content = content.replace("sigmaSet", str(sigmaSet))
    content = content.replace("cutSet", str(cutSet))
    content = content.replace("VCCFileNameSet", str(VCCFileNameSet))
    
    # Write the modified content to the output file
    with open(output_filename, 'w') as file:
        file.write(content)

from datetime import datetime

def get_current_time_formatted():
    return datetime.now().strftime("%Y-%m-%d-%H-%M-%S")

exportFileName = f"impactResults_{get_current_time_formatted()}.json"
```

```python
history = pd.DataFrame()


for config in ["iris", "RPP"]:

    # http://physics-elog.slac.stanford.edu/facetelog/show.jsp?dir=/2024/41/07.10&pos=2024-10-07T18:12:54
    # Uncomment the relevant section in the _template file to pick between a radial distribution or a VCC image file
    if config == "iris":
        sigmaSet = 1.78
        cutSet = 2.67
        VCCFileNameSet = "2024-10-07_VCC_iris"

    if config == "RPP":
        sigmaSet = 1.48
        cutSet = 2.52
        VCCFileNameSet = "2024-10-07_VCC_RPP"

    
        

    for solSet in np.arange(-0.38, -0.44, -0.01):
        
        #for MTESet in [1, 1e2, 1e4, 1e6]:
        for MTESet in [100, 400, 1000]:
    
            makeDistgenFile(
                MTESet = MTESet,
                sigmaSet = sigmaSet,
                cutSet = cutSet, 
                VCCFileNameSet = VCCFileNameSet
                )
            
            savedData = {}
            
            print(f"MTESet: {MTESet}")
            print(f"Solenoid: {solSet}")
            
            I = runImpact(
                gridCount = 32,
                numMacroParticles = 1e6,
                GFILESuffix = '2024-10-07_distgen.yaml',
                L0APhaseOffset = 0,
                solenoidTValue = solSet,
                returnImpactObject = True
            )
        
            
            #P = I.particles['PR10241'].copy()
            P = I.particles['L0AFEND'].copy()
            
            display(plotMod(P, 'x', 'y',  bins=100))
            display(plotMod(P, 'x', 'xp', bins=100))

            

            savedData["config"] = config
            savedData["sigmaSet"] = sigmaSet
            savedData["cutSet"] = cutSet
            
            savedData["MTESet"] = MTESet
            savedData["solSet"] = solSet
    
            #This actually seems survive export to JSON. Makes big files though
            #savedData["P"] = P 
            
            savedData["norm_emit_x"] = P.norm_emit_x
            savedData["norm_emit_y"] = P.norm_emit_y
        
            savedData["emitSI90_x"] = smallestIntervalImpliedEmittance(P, plane = "x", percentage = 0.90)
            savedData["emitSI90_y"] = smallestIntervalImpliedEmittance(P, plane = "y", percentage = 0.90)
        
        
            savedData[f"sigmaSI90_x"] = smallestIntervalImpliedSigma(P.x, percentage = 0.90)
            savedData[f"sigmaSI90_y"] = smallestIntervalImpliedSigma(P.y, percentage = 0.90)
        
            # print(f"""{savedData["norm_emit_x"]}, {savedData["norm_emit_y"]}""")
            # print(f"""{savedData["emitSI90_x"]}, {savedData["emitSI90_y"]}""")

            print(savedData)
            
            tmpData = pd.Series( savedData ) 
            history = pd.concat([history, tmpData.to_frame().T])
    
            history.to_json(exportFileName, orient='records')
```
